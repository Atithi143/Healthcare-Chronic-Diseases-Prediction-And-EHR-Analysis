```python
import pandas as pd
```

## Step 1: Load the Dataset

```python
df = pd.read_csv("/content/ckd.csv")
```

### **Step 2: View the First 5 Rows**

```python
df.head()
```

## Step 3: Check Dataset Information

```python
df.info()
```

## Step 4: Check Dataset Shape

```python
df.shape
```

## Step 5: Check Missing Values

```python
df.isnull().sum()
```

## Step 6: Check Duplicate Rows

```python
df.duplicated().sum()
```

## Step 7: Remove the ID Column

```python
if 'id' in df.columns:
    df.drop("id", axis=1, inplace=True)
```

## Step 8: Check Data **Types**

```python
df.dtypes
```

## Step 11: Fill Missing Numerical Values

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

# Step 12: Fill Missing Categorical **Values**

```python
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# **Step** 13: Remove ExtraSpaces

```python
for col in cat_cols:
    df[col] = df[col].str.strip()
```

# **Step 14: Check Missing Values Again**

```python
df.isnull().sum()
```

# **Step 15: View the Cleaned Dataset**

```python
df.head()
```

## **Step 16: Save the Cleaned Dataset**

```python
df.to_csv("/content/ckd.csv", index=False)
```
